# **Multiple_Linear_Regression**

A multible linear regression model represents the relationship between <u>Two or More Independent variables</u> and <u>One dependent Variable</u>

Equation:Y=a1X1+a2X2-a3X3+b
independant variables: X1,X2,X3
dependent variables: Y
Coefficients/slops: a1,a2,a3,
interceptor/constatnt: b

```python
#To convert the column into categorical columns

states = pd.get_dummies(X['State'])
states
```

```python
# Drop the state coulmn
X=X.drop('State', axis=1)  # we will replace this column with 0,1

# concat the dummy variables to our dataset
X=pd.concat([X,states], axis=1)
```

```python
# Display the resulted dataset
X
```

# **!** We have a **problem**
We can see that we have NULL values with the new data

So, we need to clean our data from null values
[**Note**: the size of dataset after `.concat` is bigger now]

```python
# Test if we have null values
X.isnull()   # this function returns True for the null values
```

```python
# How many Null value?
X.isnull().sum()
```

```python
# Now let's clean dataset from null values
X.dropna()
```

```python
# As you see we deleted all data , in other cases we may delet 50% of our data and so on,
# to avouid this problem we may replace NULL values with other values

X.fillna(0)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# call training model
from sklearn.linear_model import LinearRegression   # for multi linear regression we use linear regression model but the number of independents are >1
model= LinearRegression()    #  create object of the learning model that you chose

# fit your training data to the model object
model.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = model.predict(X_test)
```

```python
# What is the RMSE of your model?

from sklearn.metrics import mean_squared_error
import numpy as np

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print('Root Mean Square Error:', rmse)
```
